```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
color = {0: '#f8766d', 1:'#00bfc4'}
```

```python
fid = pd.read_csv('fid.csv'); fid.head()
```

```python
plt.figure(figsize=(10, 7))
steps = range(1000, 11000, 1000)
idx_min = fid.fid.values.argmin()
plt.plot(steps, fid.fid, ':+', markersize=10)
plt.axvline(x=steps[idx_min],
            lw=1.5,
            color='r',
            label=f'Best FID: {fid.fid.values[idx_min]:0.2f}')
plt.ylabel('FID', fontsize=15)
plt.xlabel('Step', fontsize=15)
leg = plt.legend(framealpha=0, loc='best', fontsize=15)
for text in leg.get_texts():
    plt.setp(text, color='r')
for item in leg.legendHandles:
    item.set_visible(False)
plt.title('Fake Horse Image Evaluation from CIFAR-10', fontsize=20)
plt.savefig('FID.png')
plt.show()
```

```python
loss = pd.read_csv('history.csv'); loss.head()
loss['steps'] = range(100, 10100, 100)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(loss.steps, loss.LossG, label='Generator Network', c=color[0])
plt.plot(loss.steps, loss.LossD, label='Discriminator Network', c=color[1])
plt.legend(frameon=False)
plt.xlabel('Step', fontsize=15)
plt.ylabel('Hinge loss', fontsize=15)
plt.title("Training History on CIFAR-10 Horse Image)", fontsize=20)
plt.savefig('train_history.png')
plt.show()
```
